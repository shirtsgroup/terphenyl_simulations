# Plumed Analysis Notebook

Using this notebook to visualize and analyze the output from the H-bond metadynamics simulations

```python
import plumed
import matplotlib.pyplot as plt
```

## Plotting Individual H-bonds over time

```python
data = plumed.read_as_pandas("COLVAR_REWEIGHT")
data
```

```python
fig, ax = plt.subplots(4,1, figsize = [15,10])
ax[0].plot(data.time*200/1000, data.values[:,1])
ax[0].set_ylabel("Hydrogen Bond 1")
ax[1].plot(data.time*200/1000, data.values[:,2])
ax[1].set_ylabel("Hydrogen Bond 2")
ax[2].plot(data.time*200/1000, data.values[:,3])
ax[2].set_ylabel("Hydrogen Bond 3")
ax[3].plot(data.time*200/1000, data.values[:,4],'r')
ax[3].set_ylabel("Bias Energy")
ax[3].set_xlabel("Time (ns)")
```
